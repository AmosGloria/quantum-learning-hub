**Introduction: Difference between Classical bits and Quantum bits**
Classical bits are the basic building blocks of classical computing, they are the smallest unit of data for classical computers which can either be 1 or 0.
Qubits are the basical building blocks of quantum computing, they can exist in a superposition of states, which means it can be 0 or 1 or both 0 and 1 simultaneously.

Meaning Hadamard Gate: The Hadamard gate denoted as H acts on a single qubit with the purpose of transforming a qubit into superposition where the qubit exist in both |1> and |0> states simultaneously.

```python
bell = QuantumCircuit(2)
bell.h(0)
bell.cx(0,1)
bell.measure_all()
bell.draw("mpl")
```
